```python
import pandas as pd
car_data=pd.read_csv("toyota.csv")
samp=car_data.copy(deep=True)
print(samp)
```

```python
samp.insert(10, "Price_Class", '')
```

```python
for i in range(len(samp["Price"])):
    if samp["Price"][i]<=8450:
        samp["Price_Class"][i]='Low'
    elif samp["Price"][i]>11950:
        samp["Price_Class"][i]='High'
    else:
        samp["Price_Class"][i]='Medium'
print(samp)
```

```python
samp["Price_Class"].value_counts()
```

```python
samp.insert(11, "Age_Converted", 0)
```

```python
samp
```

```python
def c_convert(val):
    val_con=val/12
    return val_con
samp["Age_Converted"]=c_convert(samp["Age"])
samp["Age_Converted"]=round(samp["Age_Converted"])
```

```python
def cal_km_per_mon(val1, val2):
    val_con=val1/12
    ratio=val2/val1
    return [round(val_con), ratio]
samp["Age_Converted"], samp["Km_per_month"]=cal_km_per_mon(samp["Age"], samp["KM"])
samp
```
