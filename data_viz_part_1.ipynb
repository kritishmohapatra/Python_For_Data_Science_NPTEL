```python
plt.scatter(car_data["Age"], car_data["Price"], c="blue")
plt.title("Scatter plot of price vs age of the cars")
plt.xlabel("Age(months)")
plt.ylabel("Price(Euros)")
```

# Histogram


```python
plt.hist(car_data["KM"])
```

```python
plt.hist(car_data["FuelType"])
```

```python
plt.hist(car_data["KM"], color='green', edgecolor="white", bins=5)
plt.title("Histogram")
plt.xlabel("KM")
plt.ylabel("Frequency")
plt.show()
```

# Bar plot

```python
counts=[979, 120, 12]
fuel=("Petrol", "Diesel", "CNG")
index=np.arange(len(fuel))
plt.bar(index, counts, color=["red", "blue", "cyan"])
plt.title("Bar plot")
plt.xlabel("Fuel types")
plt.ylabel("Freq")
plt.xticks(index, fuel, rotation=50)
plt.show()
```
