```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
car_data=pd.read_csv("toyota.csv",  na_values=["??", "???"])
```

```python
car_data.dropna(axis=0, inplace=True)
```

```python
car_data
```

# Scatter plot

```python
sns.set(style="darkgrid")
sns.regplot(x=car_data["Age"], y=car_data["Price"])
```

```python
sns.set(style="darkgrid")
sns.regplot(x=car_data["Age"], y=car_data["Price"], fit_reg=False)
```

```python
sns.set(style="darkgrid")
sns.regplot(x=car_data["Age"], y=car_data["Price"],marker="*", fit_reg=False)
```

```python
sns.lmplot(x="Age", y="Price", data=car_data, fit_reg=False, hue="FuelType", legend=True, palette="Set1")
```

# Histogram

```python
sns.displot(car_data["Age"])
```

```python
sns.displot(car_data["Age"], kde=True)
```

```python
sns.displot(car_data["Age"], bins=5)
```

# Bar plot

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create subplots with specified height ratios
fig, (ax_b, ax_h) = plt.subplots(2, figsize=(8, 6), gridspec_kw={"height_ratios": (.15, .85)})

# Boxplot on top
sns.boxplot(x=car_data["Price"], ax=ax_b)

# Histogram below (use sns.histplot instead of sns.displot for Axes compatibility)
sns.histplot(car_data["Price"], ax=ax_h, kde=False)

# Improve layout
plt.tight_layout()
plt.show()
```

# Pair Wise

```python
sns.pairplot(car_data, kind="scatter", hue="FuelType")
plt.show()
```
