```python
import pandas as pd
car_data=pd.read_csv('toyota.csv', index_col=0, na_values=["??", "???"])
```

```python
samp=car_data.copy()
```

```python
samp
```

# Frequency Tables

# crosstab

```python
pd.crosstab(index=samp["FuelType"], columns="count", dropna=True)
```

# Two way Tables

```python
pd.crosstab(index=samp["Automatic"], columns=samp["FuelType"], dropna=True)
```

# Two way table-joint probability

```python
pd.crosstab(index=samp["Automatic"], columns=samp["FuelType"],normalize=True, dropna=True)
```

# Mariginal probability"

```python
pd.crosstab(index=samp["Automatic"], columns=samp["FuelType"],normalize=True, dropna=True, margins=True)
```

# Conditional Probability

```python
pd.crosstab(index=samp["Automatic"], columns=samp["FuelType"],normalize="index", dropna=True, margins=True)
```

```python
pd.crosstab(index=samp["Automatic"], columns=samp["FuelType"],normalize="columns", dropna=True, margins=True)
```
